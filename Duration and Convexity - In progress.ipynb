```python
import numpy as np
import chart_studio
import sympy as sym
import seaborn as sns
import plotly.graph_objs as go
import matplotlib.pyplot as plt
import chart_studio.plotly as py

from sympy import *

sns.set_style("white")
chart_studio.tools.set_config_file(world_readable = True, sharing = 'public')
```

```python
d, r, t = sym.symbols('d r t')
```

```python
class style:
   BOLD = '\033[1m'
   END = '\033[0m'
```

```python
f0 = 1 / ((1+r/2)**(t))
f20 = 1 / ((1+r/2)**(40))

print(style.BOLD + 'Preço do título quando t = 0:' + style.END)
display(f0)
print(style.BOLD + 'Preço do título quando t = 20:' + style.END)
display(f20)
print(style.BOLD + 'Dólar Duration quando t = 20:' + style.END)
display(sym.diff(f20, r, 1))
print(style.BOLD + 'Dólar Convexity quando t = 20:' + style.END)
display(sym.diff(f20, r, 2))
```

```python
fdiff = sym.diff(f20, r, 1)

fx = lambdify(r, f20, modules = ['numpy'])

fxdiff = lambdify(r, fdiff, modules = ['numpy'])

xvals = np.linspace(0, 10, 20)
```

```python
plt.plot(xvals, fx(xvals), color = 'blue')
plt.plot(xvals, fxdiff(xvals), color = 'red')
```
